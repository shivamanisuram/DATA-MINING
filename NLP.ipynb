```python
import re
import pandas as pd
```

```python
df=pd.read_csv("/content/Consumer_Complaints.csv")
```

```python
print("Columns in dataset:", df.columns.tolist())
```

```python
df.info()
```

```python
text_column = "Consumer Complaint"
```

```python
phone_pattern = r'(?:\+?\d{1,3}[-.\s]?)?(?:\(?\d{2,4}\)?[-.\s]?)?\d{3,4}[-.\s]?\d{4}'
```

```python
def find_phone_numbers(text):
    phones = re.findall(phone_pattern, str(text))
    phones = [p.strip() for p in phones if p.strip()]
    return ", ".join(phones) if phones else "(none)"
```

```python
def clean_text(text):
    text = str(text)
    text = re.sub(phone_pattern, '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'[^A-Za-z0-9\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
df["phones_found"] = df[text_column].apply(find_phone_numbers)
df["cleaned_text"] = df[text_column].apply(clean_text)
```
